## Create a Function that show the Equity and Amount per Shark

```python
def sharks(data):
    list=["ashneer_deal","anupam_deal","aman_deal","namita_deal","vineeta_deal","peyush_deal","ghazal_deal"]
    for i in list:
        deal=data[["amount_per_shark","equity_per_shark"]][data[i]==1]
        print("\n{} Deal with {} ".format(len(deal),i[:-5]))
        print(deal)
```

## Ashneer Deals

```python
ash_grover = df[df["ashneer_deal"]==1]
ash_grover
```

```python
sharks(ash_grover)
```
