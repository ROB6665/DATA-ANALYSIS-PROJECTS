# How to Fill NAN VALUES / MISSING VALUES in data ?

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
ds={"Name":["rob","ash","jat","aml","rhl","ayn","jtu","glu"],
    "Age":[24,26,21,np.nan,np.nan,170,np.nan,20],
    "Height":[5.9,5.2,6.1,np.nan,np.nan,np.nan,5.7,5.5],
    "Gender":["Male","Male",np.nan,np.nan,"Male","Male","Male","Female"]}
```

```python
df=pd.DataFrame(ds)
df
```

```python
print("------MISSING VALUES------")
print("In Count")
print(df.isnull().sum())
print("******************************")
print("In Percentage")
print(df.isnull().sum()/df.shape[0]*100)
```

```python
print("_______Information______")
print(df["Age"].mean())
print(df["Age"].median())
print("******************")
print(df["Height"].mean())
print(df["Height"].median())
print("******************")
df["Gender"].mode()
```

```python
df["Age"]=df["Age"].fillna(df["Age"].median())
df["Height"]=df["Height"].fillna(df["Height"].mean())
df["Gender"]=df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
print("_______Information______")
print(df["Age"].mean())
print(df["Age"].median())
print("******************")
print(df["Height"].mean())
print(df["Height"].median())
print("******************")
df["Gender"].mode()
```

```python
print("------MISSING VALUES------")
print("In Count")
print(df.isnull().sum())
print("******************************")
print("In Percentage")
print(df.isnull().sum()/df.shape[0]*100)
```
