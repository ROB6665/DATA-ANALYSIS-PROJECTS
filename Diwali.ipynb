# Exploratory Data Analysis

```python
df.columns
```

## Gender

```python
sns.countplot(x="Gender",data=df);
```

```python
ax = sns.countplot(x="Gender",data=df);
for bars in ax.containers:
    ax.bar_label(bars)

# This method is to add data label
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_gen
```

```python
sns.barplot(x = "Gender", y = "Amount", data = sales_gen);
```

From above graphs we can see that most of the buyers are female and even the purchasing of females greater than male customers

## Age

```python
sns.countplot(x = "Age Group", data=df);
```

```python
ax = sns.countplot(x = "Age Group", hue = "Gender", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount VS Age group

sales_age = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_age
```

```python
sns.barplot(x = "Age Group", y = "Amount", data = sales_age);
```

From above graphs we can see that most of the buyers are of age group between 26 - 35 yrs female

## State

```python
# Total number of order from top 10 states

sales_state = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)
sales_state
```

```python
sns.set(rc={"figure.figsize":(16,6)})
sns.barplot(data = sales_state, x= "State", y= "Orders");
plt.xticks(rotation=30);
```

From above chart we can see that unexpectedly most of the orders are from Uttar pardesh, Maharashtra, karnataka, Delhi and then Madhya pardesh.

```python
amt_state = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(5)
amt_state
```

```python
sns.set(rc={"figure.figsize":(16,6)})
sns.barplot(data = amt_state, x= "State", y= "Amount");
plt.xticks(rotation=30);
```

From above chart we can see that these are the top 5 states which have highest sales amount

## Marital status

```python
sns.set(rc={"figure.figsize":(6,5)})
ax = sns.countplot(data = df, x="Marital_Status")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending= False)

sns.set(rc={"figure.figsize":(12,6)})
sns.barplot(x="Marital_Status",y="Amount",hue="Gender", data=sales_state);
```

## Occupation

```python
sns.set(rc={"figure.figsize":(20,8)})
ax = sns.countplot(x="Occupation",data = df)
plt.xticks(rotation=30);
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(20,8)})
sns.barplot(x="Occupation",y="Amount",data=sales_state);
```

From above graph we can see that the most of the buyers are working in IT, Aviation and Health care sector

## Product category 

```python
sns.set(rc={"figure.figsize":(24,8)})
ax = sns.countplot(data = df, x="Product_Category")
plt.xticks(rotation=30);
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,8)})
sns.barplot(x="Product_Category", y="Amount",data = sales_state);
```

From above graph we can see that most of the sold products are from food, clothing and electronics category

```python
sales_state = df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,6)})
sns.barplot(data = sales_state, x="Product_ID",y= "Orders");
```

```python
fig1,ax1 = plt.subplots(figsize=(12,7));
df.groupby("Product_ID")["Orders"].sum().nlargest(10).sort_values(ascending=False).plot(kind="bar");
```
